# Example notebook : structured dataset variable plots and saving image

## Load packages

```python
from transparentai.datasets import StructuredDataset, load_adult

# If you want to save different plots that the package generate
import transparentai.plots as plots
# save_plot allows to save
plots.save_plot = True

# For a different directory than the current
plots.save_dir = 'plots/'
# If you want to see details
plots.verbose = 1
```

## Load adult dataset

```python
adult = load_adult()
```

## Create StructuredDataset object

```python
target = 'income'
dataset = StructuredDataset(df=adult, target=target)
```

## Use different plot function

```python
dataset.plot_one_categorical_variable(var='gender')
```

```python
dataset.plot_one_numeric_variable(var='age')
```

```python
dataset.plot_one_cat_and_num_variables(var1='gender', var2='age')
```

```python
dataset.plot_two_numeric_variables(var1='age', var2='hours-per-week')
```
