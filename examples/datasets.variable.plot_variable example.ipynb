`plot_variable` example
-----

### Load packages

```python
import numpy as np
import pandas as pd
import random
import datetime

from transparentai.datasets import load_adult
from transparentai.datasets import variable
```

### Load data

```python
data = load_adult()
```

Add a date variable and replace `?` by nan values. 

```python
def generate_birthdate(age):
    month = random.randint(1,12)
    day = random.randint(1,28)
    year = 2020 - age
    return datetime.date(year, month, day)

data['birthdate'] = data['age'].apply(generate_birthdate)
data = data.replace('?',np.nan)
```

```python
data.sample(3)
```

### Use `plot_variable` function

`age` variable

```python
variable.plot_variable(data['age'])
# variable.plot_variable(data['age'], legend=data['income'])
```

`capital-loss` variable

```python
# variable.plot_variable(data['capital-loss'], ylog=True)
variable.plot_variable(data['capital-loss'], legend=data['income'], ylog=True)
```

`workclass` variable

```python
variable.plot_variable(data['workclass'])
# variable.plot_variable(data['workclass'], legend=data['income'])
```

`birthdate` variable

```python
# variable.plot_variable(data['birthdate'])
variable.plot_variable(data['birthdate'], legend=data['income'])
```

So this is some examples using `variable.plot_variable` function.
