# Example notebook : analyse structured dataset for a binary classification

## Load packages

```python
from transparentai.datasets import StructuredDataset, load_boston
```

## Load boston dataset

```python
boston = load_boston()
```

## Create StructuredDataset object

```python
dataset = StructuredDataset(df=boston)
```

## Overview

```python
dataset.plot_dataset_overview()
```

## Missing values

```python
dataset.plot_missing_values()
```

## Each variable (one by one)

```python
dataset.plot_variables()
```

## Numeric variables relationship

```python
dataset.plot_numeric_var_relation()
```

## Numeric and categorical variables relationship

```python
dataset.plot_cat_and_num_variables()
```

## Correlation

```python
dataset.plot_correlations()
```
