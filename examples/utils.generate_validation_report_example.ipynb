`utils.generate_validation_report` example for binary classifier
----

### Load package

```python
from transparentai.datasets import load_adult
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

from transparentai import utils
```

### Load data

```python
data = load_adult()
X, Y = data.drop(columns='income'), data['income']
```

```python
X = X.select_dtypes('number')
Y = Y.replace({'>50K':1, '<=50K':0})
```

### Split train test

```python
X_train, X_valid, y_true, y_true_valid = train_test_split(X, Y, test_size=0.33, random_state=42)
```

### Train model

```python
clf = RandomForestClassifier()
clf.fit(X_train,y_true)
```

### Generate report 

**This function generate an output pdf file : [validation_report.pdf](out/validation_report.pdf).**

```python
utils.generate_validation_report(clf, X_train, y_true, X_valid, y_true_valid, out='validation_report.pdf')
```
