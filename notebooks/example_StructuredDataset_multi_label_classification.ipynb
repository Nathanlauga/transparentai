# Example notebook : analyse structured dataset for a multi labels classification

## Load packages

```python
from transparentai.datasets import StructuredDataset, load_iris
```

## Load iris dataset

```python
iris = load_iris()
```

## Create StructuredDataset object

```python
target = 'iris plant'
dataset = StructuredDataset(df=iris, target=target)
```

## Overview

```python
dataset.plot_dataset_overview()
```

## Missing values

```python
dataset.plot_missing_values()
```

## Each variable (one by one)

```python
dataset.plot_variables()
```

## Numeric variables relationship

```python
dataset.plot_numeric_var_relation()
```

## Numeric and categorical variables relationship

```python
dataset.plot_cat_and_num_variables()
```

## Correlation

```python
dataset.plot_correlations()
```
