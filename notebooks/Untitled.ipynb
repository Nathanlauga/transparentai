# Analyse iris dataset

## Load packages

```python
import pandas as pd
from IPython.display import display, Markdown

# transparentai package : https://github.com/Nathanlauga/transparentai
import transparentai.explore as explore
from transparentai.utils import remove_var_with_one_value
from transparentai import datasets
```

## Load dataset

```python
iris = datasets.load_iris()
```

```python
target = 'iris plant'
```

## Quick overview

```python
display(Markdown(f'#### {iris.shape}'))
display(iris.head())
```

## Analyse : missing values

```python
display(Markdown('#### Missing values for adult dataset'))
explore.show_missing_values(iris)
```

## Analyse : each variable

```python
dataset = remove_var_with_one_value(iris)
```

```python
explore.show_df_vars(df=iris, target=target)
```

## Analyse : numericals variables relations

```python
explore.show_df_numerical_relations(df=iris, target=target)
```

## Analyse : numericals & categoricals variables relations

```python
explore.show_df_num_cat_relations(df=iris, target=target)
```

## Analyse variables correlations

```python
explore.show_df_correlations(df=iris)
```
